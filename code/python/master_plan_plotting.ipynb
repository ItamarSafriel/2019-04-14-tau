# Basic Plotting

Generate data

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y)
```

Let's make points

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y, 's', color='r')
```

### Log axes

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y, 's')
ax.semilogx()
```

### Scale Everything

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y, 's')
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Random Points')
```

# Scatter Plots
<a href = "https://matplotlib.org/gallery/shapes_and_collections/scatter.html#sphx-glr-gallery-shapes-and-collections-scatter-py"> Source: Matplotlib example </a>

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.scatter(x, y)
plt.show()
```

### color code

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
colors = np.random.rand(N) #values between 0--> represent values on a color map

ax.scatter(x, y,c=colors)
plt.show()
```

### Scale

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

ax.scatter(x, y, s=area, c=colors)
plt.show()
```

### Change opacity

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

# Errorbars

generate data

```python
x = np.arange(10)+np.random.randn(10)
y = np.arange(10)+np.random.randn(10)
xerr = np.random.randn(10)
yerr = np.random.randn(10)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.errorbar(x,y, yerr=yerr)
```

### Let's clean that up

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.errorbar(x,y, yerr=yerr, fmt='o')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.errorbar(x,y, yerr=yerr, fmt='o', capsize=2)
```

### Add x errors

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.errorbar(x,y, yerr=yerr, xerr=xerr, fmt='o', capsize=2)
```

# Legend

# generate data

```python
x = np.arange(10)+np.random.randn(10)
y = x+np.random.randn(10)
y2 = x**2 + np.random.randn(10)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, y, 'o', label='linear')
ax.plot(x, y2, 's', label='quadratic')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
l1, = ax.plot(x, y, 'o', label='linear')
l2, = ax.plot(x, y2, 's', label='quadratic')
ax.legend([l1], ['linear only'])
```

# Show an Image

```python
# Generate fake data
np.random.seed(0)
y, x = np.mgrid[:1024, :1024]
z = 2. * x ** 2 - 0.5 * x ** 2 + 1.5 * x * y - 1.
z += np.random.normal(0., 0.1, z.shape) * 50000.
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.imshow(z)
```

### Change the color map

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
im = ax.imshow(z, cmap='bone')
```

### Change the contrast limits

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.imshow(z, cmap='bone', vmin=-20000, vmax=1E6)
```

### Implement DS9 zscale

```python
from astropy.visualization import ZScaleInterval

def zscale(img):
    zscale = ZScaleInterval()
    vmin, vmax = zscale.get_limits(img)
    return vmin, vmax
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
zscale_vmin, zscale_vmax = zscale(z)
ax.imshow(z, cmap='bone', vmin=zscale_vmin, vmax=zscale_vmax)
```

### Add a colorbar

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
zscale_vmin, zscale_vmax = zscale(z)
im = ax.imshow(z, cmap='bone', vmin=zscale_vmin, vmax=zscale_vmax)
fig.colorbar(im)
```
